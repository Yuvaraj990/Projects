In Univariate Analysis for Numerical Variable we need to look for the Distributions whether it is Normally distributed or it is skewed and also identify the outliers present in the Data.The graphs we used above are Box-plot,Histogram,Q-Q plot.
* Box-Plot :
   * A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”). It can tell you about your outliers and what their values are. It can also tell you if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed.

* Histogram :
   * A histogram is a graphical representation of a grouped frequency distribution with continuous classes. It is an area diagram and can be defined as a set of rectangles with bases along with the intervals between class boundaries and with areas proportional to frequencies in the corresponding classes. In such representations, all the rectangles are adjacent since the base covers the intervals between class boundaries. The heights of rectangles are proportional to corresponding frequencies of similar classes and for different classes, the heights will be proportional to corresponding frequency densities.

* Q-Q plot :
    * The Quantile-Quantile  plot is similar to a probability plot. For a probability plot, the quantiles for one of the data samples are replaced with the quantiles of a theoretical distribution

In Bi-Variate Analysis if two variables are categorical then we use cross-tabulation i.e., contingency-table , Heat-Map , clustered column chart.
* Heat-Map :
  * A heat map is a two-dimensional representation of data in which values are represented by colors.

* Clustered-column-Chart :
  * A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally. A vertical bar chart is sometimes called a column chart.
  *A clustered column chart displays more than one data series in clustered vertical columns. Each data series shares the same axis labels, so vertical bars are grouped by category.  Clustered columns allow the direct comparison of multiple series, but they become visually complex quickly. They work best in situations where data points are limited.

```python
# Age vs Heart-Disease vs Gender :
# Two categorical variables and one Numeric Variable : 
# we can see the Numeric variable Mean and Median values across the two categorical variables.

sns.set_style('white')

plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
fig=sns.barplot(data=df,x='sex',y='age',hue='target',estimator=np.mean)
fig.bar_label(fig.containers[0])
fig.bar_label(fig.containers[1])
plt.title('Mean Values of Age across Gender vs Heart-Disease')
plt.xlabel('Gender')
plt.xticks([0,1],['Female','Male'])
plt.ylabel('Age')

plt.subplot(2,2,2)
fig=sns.barplot(data=df,x='sex',y='age',hue='target',estimator=np.median)
fig.bar_label(fig.containers[0])
fig.bar_label(fig.containers[1])
plt.title('Median Values of Age across Gender vs Heart-Disease')
plt.xlabel('Gender')
plt.xticks([0,1],['Female','Male'])
plt.ylabel('')

plt.subplot(2,2,3)
sns.heatmap(df[['age','sex','target']].groupby(by=['sex','target']).mean(),annot=True,fmt='f',cmap='turbo')
plt.ylabel('Gender-Disease')
plt.xlabel('Age Mean Values')

plt.subplot(2,2,4)
sns.heatmap(df[['age','sex','target']].groupby(by=['sex','target']).median(),annot=True,fmt='f',cmap='turbo')
plt.ylabel('')
plt.xlabel('Age Median Values')


plt.subplots_adjust(right=1.2)
plt.show()
```

```python
# thal vs chol vs target :  Thalassemia vs Cholestoral vs Heart-Disease 

plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
fig=sns.barplot(data=df,x='thal',y='chol',hue='target',estimator=np.mean)
fig.bar_label(fig.containers[0])
fig.bar_label(fig.containers[1])
plt.xlabel('Thalassemia Levels')
plt.ylabel('Cholestoral')
plt.title('Mean Values of Cholestoral across Thalassemia vs Heart-Disease')


plt.subplot(2,2,2)
fig=sns.barplot(data=df,x='thal',y='chol',hue='target',estimator=np.median)
fig.bar_label(fig.containers[0])
fig.bar_label(fig.containers[1])
plt.title('Median Values of Cholestoral across Thalassemia vs Heart-Disease')


plt.subplot(2,2,3)
sns.heatmap(df[['thal','chol','target']].groupby(by=['thal','target']).mean(),annot=True,fmt='f',cmap='viridis')
plt.ylabel('Thalassemia-Disease')
plt.xlabel('Cholestoral Mean Values')

plt.subplot(2,2,4)
sns.heatmap(df[['thal','chol','target']].groupby(by=['thal','target']).median(),annot=True,fmt='f',cmap='viridis')
plt.ylabel('')
plt.xlabel('Cholestoral Median Values')


plt.subplots_adjust(right=1.2)
plt.show()
```
